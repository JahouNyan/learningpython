```python
import pandas as pd
```

```python
df = pd.read_json("RedditAPItest.json")
```

```python
#Show a histogram with the scores,separated in 10 bins.

score = df.score
score.plot(kind = 'hist', bins = 10)
```

```python
#* B) Show a horizontal bar chart with the five highest scores.
score = score.sort_values(ascending=False).head()
score.plot(kind = 'bar')
```

```python
score2 = [{"total": score.sum()-score.max(),
          "highscore":score.max()}]
df2 = pd.DataFrame(score2, index = ["highscore", "total"])
df2.plot.pie(y = "total")
```
